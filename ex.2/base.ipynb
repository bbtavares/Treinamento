```python
import pandas as pd
import numpy as np
import csv
```

```python
database_1 = pd.read_csv(rf'C:\Users\Beatriz Brito\Desktop\CEPEDI\ex.2\datasets\WorldCupMatches.csv', 
    low_memory=False).dropna(axis=0)
database_1.columns = database_1.columns.str.replace(' ', '_').str.lower()
database_1 = database_1.drop(columns=[
    'datetime',
    'win_conditions',
    'win_conditions',
    'attendance',
    'half-time_home_goals',
    'half-time_away_goals',
    'referee',
    'assistant_1',
    'assistant_2',
    'roundid',
    'matchid',
    'home_team_initials',
    'away_team_initials'
])
database_1['year'] = database_1['year'].astype('int64')
database_1['stage'] = database_1['stage'].astype('category')
database_1['stadium'] = database_1['stadium'].astype('string')
database_1['city'] = database_1['city'].astype('string')
database_1['home_team_name'] = database_1['home_team_name'].astype('string')
database_1['away_team_name'] = database_1['away_team_name'].astype('string')
database_1['home_team_goals'] = database_1['home_team_goals'].astype('int32')
database_1['away_team_goals'] = database_1['away_team_goals'].astype('int32')

database_2 = pd.read_csv(rf'C:\Users\Beatriz Brito\Desktop\CEPEDI\ex.2\datasets\WorldCups.csv', 
    low_memory=False)
database_2.columns = database_2.columns.str.replace(' ', '_').str.lower()
database_2['goals_scored'] = database_2['goalsscored']
database_2['qualified_teams'] = database_2['qualifiedteams']
database_2['matches_played'] = database_2['matchesplayed']
database_2['audience'] = database_2['attendance']
database_2 = database_2.drop(columns=[
    'goalsscored',
    'qualifiedteams',
    'matchesplayed',
    'attendance'
])
database_2['country'] = database_2['country'].astype("string")
database_2['winner'] = database_2['winner'].astype("string")
database_2['runners-up'] = database_2['runners-up'].astype("string")
database_2['third'] = database_2['third'].astype("string")
database_2['fourth'] = database_2['fourth'].astype("string")
database_2['audience'] = database_2['audience'].str.replace('.', '')
database_2['audience'] = pd.to_numeric(database_2['audience'])
```

```python
# Cadastro da loja a

cadastro_a = {'Id': ['A1A2A3A','B1B2B3B','C1C2C3C','D1D2D3D','E1E2E3E','F1F2F3F'],
'Nome': ['Andre', 'Breno', 'Joao', 'Caio', 'Pedro', 'Bia'],
'Idade': [20,35,27,18,29,25],
'CEP': ['11111-111','22222-222','33333-333','444444-444','55555-55555','66666-666']
}
loja_1 = pd.DataFrame(cadastro_a, columns = ['Id','Nome','Idade','CEP'])

# Cadastro da loja b

cadastro_b = {'Id': ['B1B2B3B','H1H2H3H','J1J2J3J','F1F2F3F','L1L2L3L'],
'Nome': ['Breno', 'Paula', 'Rafaela', 'Bia', 'Mari'],
'Idade': [39,20,21,19,20],
'CEP': ['99999-999','77777-777','55555-333','444444-111','55544-55115']
}
loja_2 = pd.DataFrame(cadastro_b, columns = ['Id','Nome','Idade','CEP'])

# Registro de compras de todas unidades

compras = {'Id': ['A1A2A3A', 'B1B2B3B', 'C1C2C3C', 'D1D2D3D', 'E1E2E3E', 'F1F2F3F', 'H1H2H3H', 'J1J2J3J', 'L1L2L3L'],
'Data': ['2021-01-01','2021-01-30','2021-02-30','2021-02-01','2021-07-20','2021-03-15','2021-11-01','2021-12-30', '2021-10-03'],
'Valor': [1000,50,460,150,30,50,500,320,100]
}
compras = pd.DataFrame(compras, columns = ['Id','Data','Valor'])
```

```python
# Clientes que compram em ambas as lojas

pd.merge(loja_1, loja_2, on=["Nome"], how="inner")
```
